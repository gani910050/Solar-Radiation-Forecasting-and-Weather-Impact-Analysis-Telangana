```
## Train-Test Split
#ts_train, exog_train → Used to train the model
#ts_test, exog_test → Used to evaluate model performance
```

```
ts = df['surface_solar_radiation_downwards_mean']
ts = ts.resample('ME').mean()
exog = exog.resample('ME').mean()
```

```
ts, exog = ts.align(exog, join='inner')
```

```
train_size = int(len(ts) * 0.8)

ts_train = ts[:train_size]
ts_test = ts[train_size:]

exog_train = exog[:train_size]
exog_test = exog[train_size:]
```

```
#Best Model Selected#
#Order = (your_best_p, your_best_d, your_best_q)
#Seasonal Order = (P, D, Q, 12)
#AIC = XXXX
```

```
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings("ignore")

candidates = [
    ((1,0,1), (0,1,1,12)),   # based on SARIMA best
    ((0,1,1), (0,1,1,12)),
    ((1,1,1), (0,1,1,12)),
    ((1,0,0), (0,1,1,12)),
    ((0,0,1), (0,1,1,12))
]

results = []

for order, seasonal_order in candidates:
    try:
        model = SARIMAX(
            ts,
            exog=exog,
            order=order,
            seasonal_order=seasonal_order,
            enforce_stationarity=False,
            enforce_invertibility=False
        )
        res = model.fit(disp=False)
        results.append([order, seasonal_order, res.aic])
    except:
        pass

# sort by AIC
results = sorted(results, key=lambda x: x[2])

for r in results:
    print(r)
```

```
#LSTM MODEL
```

```
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```
features = [
    '2m_temperature_mean',
    'total_cloud_cover_mean',
    '2m_dewpoint_temperature_mean',
    'utci_mean',
    'surface_solar_radiation_downwards_mean'
]

data = df[features]
```

```
data
```

```
data.shape
```

```
data = data[features]

print(data.shape)
```

```
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
```

```
def create_sequences(data, n_steps=30):
    X, y = [], []
    for i in range(n_steps, len(data)):
        X.append(data[i-n_steps:i])
        y.append(data[i, -1])
    return np.array(X), np.array(y)

n_steps = 30
X, y = create_sequences(scaled_data, n_steps)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```
train_size = int(len(X) * 0.8)

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(n_steps, X.shape[2])),
    Dropout(0.3),
    LSTM(64),
    Dropout(0.3),
    Dense(32),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.summary()
```

```
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    epochs=80,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[early_stop]
)
```

```
y_pred = model.predict(X_test)
```

```
n_features = scaled_data.shape[1]

dummy_pred = np.zeros((len(y_pred), n_features))
dummy_test = np.zeros((len(y_test), n_features))

dummy_pred[:, -1] = y_pred[:, 0]
dummy_test[:, -1] = y_test

y_pred_actual = scaler.inverse_transform(dummy_pred)[:, -1]
y_test_actual = scaler.inverse_transform(dummy_test)[:, -1]
```

```
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae_lstm = mean_absolute_error(y_test_actual, y_pred_actual)
rmse_lstm = np.sqrt(mean_squared_error(y_test_actual, y_pred_actual))

print("LSTM MAE:", mae_lstm)
print("LSTM RMSE:", rmse_lstm)
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(y_test_actual, label='Actual')
plt.plot(y_pred_actual, label='LSTM Predicted')
plt.legend()
plt.title("LSTM Prediction")
plt.show()
```

```
last_seq = scaled_data[-n_steps:]
current_seq = last_seq.reshape(1, n_steps, X.shape[2])

future_preds = []

for _ in range(12):
    pred = model.predict(current_seq, verbose=0)[0, 0]
    future_preds.append(pred)

    new_row = current_seq[0, -1].copy()
    new_row[-1] = pred

    current_seq = np.append(current_seq[:, 1:, :], [[new_row]], axis=1)
```

```
n_features = scaled_data.shape[1]

future_dummy = np.zeros((12, n_features))
future_dummy[:, -1] = future_preds

future_pred_actual = scaler.inverse_transform(future_dummy)[:, -1]

print("Future forecast values:")
print(future_pred_actual)
```

```
future_index = pd.date_range(
    start=data.index[-1] + pd.offsets.MonthEnd(1),
    periods=12,
    freq='M'
)

future_df = pd.DataFrame({
    'LSTM_Forecast': future_pred_actual
}, index=future_index)

print(future_df)
```

```
plt.figure(figsize=(12,5))
plt.plot(data.index, data['surface_solar_radiation_downwards_mean'], label='Historical')
plt.plot(future_df.index, future_df['LSTM_Forecast'], label='LSTM Forecast')
plt.legend()
plt.title("LSTM Future Forecast")
plt.show()
```
